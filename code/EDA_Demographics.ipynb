## Univariate and Bivariate Analyses
___

```python
#Examine total student enrollment for the 5 school year ranges
total_enrollment_df = df[["Year","Total Enrollment"]]
total_enrollment_df 
```

```python
#Visualize the total student enrollment df
plt.figure(figsize=(20, 15))
total_enrollment_df.plot.line(x="Year", y="Total Enrollment")

#Remove scientific notation from y axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False) 
plt.grid()
plt.title("\nTotal Enrollment in NYC Public Schools 2015 - 2020\n", fontsize=16, weight='bold', style='italic')
plt.xlabel("\nYear\n", fontsize=14, weight='bold')
plt.ylabel("\nTotal Enrollment\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

___
> **Insights on Total Enrollment**

- Citywide K-12 enrollment was highest during the 2016-2017 school year
- The lowest total K-12 enrollment academic year was 2018-2019 with 1,126,501 students
- Citywide enrollment during the 2019-2020 school year - 1,131,868 students - increased from the previous year but did not meet the 2015-2016 headcount of 1,133,963 students. This is greatly due to the impact of the Covid-19 Pandemic.

**According to the [NYC Department of Education](https://teachnyc.net/want-to-be-teacher/early-childhood-education/all-about), 3K refers to free full and half day education for three year olds in the city.**

**[PK](https://www.schools.nyc.gov/enrollment/enroll-grade-by-grade/pre-k) refers to Pre-K/Pre-Kindergarten education for 4 year olds.**
___

```python
#Examine enrollment for 3K to PK students citywide
PK_3K_total_enrollment_df = df[["Year","Grade 3K+PK (Half Day & Full Day)"]]
PK_3K_total_enrollment_df 
```

```python
#Visualize the total 3K and PK enrollment df
plt.figure(figsize=(20, 18))
PK_3K_total_enrollment_df.plot.line(x="Year", y="Grade 3K+PK (Half Day & Full Day)")

#Remove scientific notation from y axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False) 
plt.grid()
plt.title("\nTotal 3K and PK Enrollment in NYC Public Schools 2015 - 2020\n", fontsize=16, weight='bold', style='italic')
plt.xlabel("\nYear\n", fontsize=14, weight='bold')
plt.ylabel("\nTotal 3K and PK Enrollment\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

___
> **Insights on Total 3K and PK Enrollment**

- Total enrollment of 3K and PK students remained relatively steady until the 2019-2020 school year when enrollment spiked to 87,480 3 and 4 year old students.

 **Otherwise, the studentâ€™s Economic Need Value is based on the percentage of families (with school-age children) in the studentâ€™s census tract whose income is below the poverty level, as estimated by the American Community Survey 5-Year estimate (2016 ACS estimates were used in calculations for 2018-19 ENI).
